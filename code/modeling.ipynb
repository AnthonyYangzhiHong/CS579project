```python
#build the network
import networkx as nx
DG=nx.scale_free_graph(5000,alpha=0.3, beta=0.3, gamma=0.4)
```

```python
#draw out degree distribution
from collections import Counter
import numpy as np

def plot_degree_dist(G, title):
    degrees = [G.out_degree(v) for v in G]
    degree_counts = Counter(degrees)
    p_k = [(degree, 1. * count / len(G.nodes())) for degree, count in degree_counts.iteritems()]
    ks = [x[0] for x in p_k]
    x_pos = ks
    plt.xlabel('$k$')
    plt.ylabel('$P(k)$')
    plt.bar(x_pos, [x[1] for x in p_k], align='center', alpha=0.4)
    plt.locator_params(nbins=10) 
    plt.title('Out Degree Distribution for %s (mean=%.2f, N=%d)' % (title, np.mean(degrees), len(G.nodes())))
plot_degree_dist(DG, 'network_model')
```

```python
#choose starting node
#any time we want to reset the network,we can run this method
nx.set_node_attributes(DG, 'choice', 'B')
nx.set_node_attributes(DG, 'spread_times', 0)
start_node=[]
for n in DG:
    out=DG.out_edges(n)
    if len(out)>=100 and len(out)<=150:#can be adjusted in different condition
        start_node.append(n)
print start_node
for n in start_node:
    DG.node[n]['choice']='A'
    DG.node[n]['spread_times']=1
```

```python
#simluate rumor spreading in each round 
#this method is used in manully simulation,we have to run it in each round
for n in DG:
    print n
    if DG.node[n]['choice']=='B':
        believed_followings=[]
        incoming=DG.in_edges(n)
        incoming_node=[]
        for edges in incoming:
            incoming_node.append(edges[0])
        print incoming_node
        for k in incoming_node:
            if DG.node[k]['choice']=='A':
                believed_followings.append(k)
        print believed_followings
        incoming_rumor=len(believed_followings)
        if len(incoming_node)!=0:
            ratio=1.*len(believed_followings)/len(incoming_node)
        #threshold
            print ratio
            if ratio>=0.6:
                DG.node[n]['choice']='A'
                print DG.node[n]['choice']
    else:
        if DG.node[n]['spread_times']>=2:
            DG.node[n]['choice']='S'
        else:
            DG.node[n]['spread_times']=DG.node[n]['spread_times']+1
```

```python
#draw network to see which node become believer
layout = nx.spring_layout(DG)
def draw_graph(graph):
    plt.figure()
    nodes = graph.nodes()
    colors = ['r' if graph.node[n]['choice'] == 'A' else 'b' if graph.node[n]['choice']=='B' else 'g' for n in graph]
    # plt.figure(figsize=(,10))
    plt.axis('off')
    nx.draw_networkx(graph, nodelist=nodes,node_size=200, linewidths=0.5,with_labels=True,
                     width=1, node_color=colors,alpha=.5,
                     pos=layout)    
draw_graph(DG)
```
